```python
import pandas as pd

survey_results_public = pd.read_csv("survey_results_public.csv")
survey_results_public.head(5)
```

```python
count_of_unuque_responder = survey_results_public["ResponseId"].nunique()
count_of_unuque_responder
```

```python
survey_results_schema = pd.read_csv("survey_results_schema.csv")
tuple_questions = survey_results_schema['qname'].unique()
list_questions = tuple_questions.tolist()
list_questions.append("ResponseId")
existing_columns = [col for col in list_questions if col in survey_results_public.columns]
print(existing_columns)
rows_dropna = survey_results_public.dropna(subset = existing_columns, inplace = False)
count_poll_pass = rows_dropna["ResponseId"].nunique()
count_poll_pass
```

```python
mean_work_exp = survey_results_public["WorkExp"].mean()
print(f"mean: {mean_work_exp}")
median_work_exp = survey_results_public["WorkExp"].median()
print(f"median: {median_work_exp}")
min_work_exp = survey_results_public["WorkExp"].min()
print(f"min: {min_work_exp}")
max_work_exp = survey_results_public["WorkExp"].max()
print(f"max: {max_work_exp}")
std_work_exp = survey_results_public["WorkExp"].std()
print(f"std: {std_work_exp}")
variance_work_exp = survey_results_public["WorkExp"].var()
print(f"variance: {variance_work_exp}")
sem_work_exp = survey_results_public["WorkExp"].sem()
print(f"sem: {sem_work_exp}")
```

```python
remote_workers = survey_results_public[survey_results_public["RemoteWork"] == 'Remote']
count_remote_workers = remote_workers["ResponseId"].nunique()
count_remote_workers
```

```python
survey_results_public['HasPython'] = survey_results_public['LanguageHaveWorkedWith'].str.contains(r'\bPython\b')
python_users = survey_results_public[survey_results_public['HasPython'] == True]
count_python_users = python_users["ResponseId"].nunique()
python_users_percent = (count_python_users / survey_results_public["ResponseId"].nunique()) * 100
python_users_percent
```

```python
survey_results_public['Online Courses'] = survey_results_public["LearnCode"].str.contains(r'\bOnline Courses or Certification\b')
online_courses_ed = survey_results_public[survey_results_public['Online Courses'] == True]
count_online_courses_ed = online_courses_ed["ResponseId"].nunique()
count_online_courses_ed
```

```python
python_users_by_countries = python_users.groupby("Country")
median_mean_python_users_by_countries_ConvertedCompYearly = python_users_by_countries["ConvertedCompYearly"].agg(["median", "mean"])
median_mean_python_users_by_countries_ConvertedCompYearly
```

```python
survey_results_public.sort_values(by = "ConvertedCompYearly", axis=0, ascending = False, inplace=True, na_position='last')
ed_level = survey_results_public[["EdLevel", "ConvertedCompYearly"]]
ed_level.head(5)
```

```python
responders_by_age = survey_results_public.groupby("Age")
responders_by_age_python_percent = responders_by_age['HasPython'].mean().mul(100)
responders_by_age_python_percent
```

```python
percentile_75_ConvertedCompYearly = survey_results_public["ConvertedCompYearly"].quantile(0.75)
is_in_percentile_75_ConvertedCompYearly = survey_results_public[survey_results_public["ConvertedCompYearly"] >= percentile_75_ConvertedCompYearly]
is_in_percentile_75_ConvertedCompYearly_remoted = is_in_percentile_75_ConvertedCompYearly[is_in_percentile_75_ConvertedCompYearly["RemoteWork"] == "Remote"]
popular_industry = is_in_percentile_75_ConvertedCompYearly_remoted.groupby("Industry")["ResponseId"].nunique().reset_index()
popular_industry.columns = ["Industry", "Count"] 
popular_industry.sort_values(by = "Count", ascending = False, inplace = True)
popular_industry.head(10)
```
